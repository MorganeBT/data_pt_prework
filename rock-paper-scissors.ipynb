```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
options = ("stone", "paper", "scissors")

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_game_nb = 3

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
max_win_nb = 2

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def getyourchoice():
    options = ("stone", "paper", "scissors")
    answer = input("Please enter your choice : {}".format(options))
    # Using recursion : 
    if answer not in options:
        print("Please replay! What is your choice?")
        return getyourchoice()        
    return answer


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins


def getmachinechoice():
    options = ("stone", "paper", "scissors")
    answer_machine = random.choice(options)
    return(answer_machine)

  
def combatsolution(machine_answer, player_answer):
    # Equality
    if (machine_answer == "stone" and player_answer == "stone") | (machine_answer == "paper" and player_answer == "paper") | (machine_answer == "scissors" and player_answer == "scissors"): 
        return 0
    # machine_answer wins
    elif (machine_answer == "stone" and player_answer == "scissors") | (machine_answer == "scissors" and player_answer == "paper") | (machine_answer == "paper" and player_answer == "stone"):
        return 1
    # Player wins
    elif (machine_answer == "stone" and player_answer == "paper") | (machine_answer == "paper" and player_answer == "scissors") | (machine_answer == "scissors" and player_answer == "stone"):
        return 2   
        
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
count = 0
for i in range(0, max_game_nb):
    answer_machine = getmachinechoice()
    answer_player = getyourchoice()
    print("The machine plays", answer_machine)
    print("The human plays", answer_player)
    combat = combatsolution(answer_machine, answer_player)
    if combat == 2:
        count+=1
        print("The player wins", count , "time")
            
    elif combat == 1:
        print("The machine wins")
       
    else:
        print("Equality")
    print (count)
    
if (count >= max_win_nb):
    print("The player wins the game!")

else:
    print("The machine wins the game!")
```
